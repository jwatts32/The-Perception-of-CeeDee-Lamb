<h2>Import libraries

```python
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sns
```

<h2>Import data from website and pass to dataframe, grabbing only the first table on the page.

```python
url_lamb = "https://www.pro-football-reference.com/players/L/LambCe00.htm"
lamb = pd.read_html(url_lamb)
#display(lamb)
lamb_2023 = lamb[0]
display(lamb_2023)
```

<h2>Attempted to subset data, but got a 'KeyError' error message

```python
lamb_2023 = lamb_2023[["Date","Tgt","Rec","Yds"]]
display(lamb_2023)
```

<h2>Since the table I'm using includes a multi-level index, I drop the 1st level to address the error message from the previous cell.

```python
lamb_2023.columns = lamb_2023.columns.droplevel(0)
display(lamb_2023)
```

<h2>When I try to subset data again, I get no error message.

```python
lamb_2023 = lamb_2023[["Date","Tgt","Rec","Yds"]]
display(lamb_2023)
```

<h2>I notice that I have 2 columns for "Yds", one for receiving and one for rushing, so I set those columns to different names using the dataframe's index

```python
lamb_2023.columns = ["Date", "Targets", "Receptions", "Rec_Yds", "Rush_Yds"]
display(lamb_2023)
```

<h2>Drop "Rush_Yds" column since I don't need it

```python
lamb_2023 = lamb_2023.drop(columns=['Rush_Yds'])
display(lamb_2023)
```

<h2>Just by looking at the records above, I can tell there is a huge difference in Targets and Rec Yds between the first half of the season and the second half.  I'm curious to see what the averages are for these 2 categories.

```python
first_6_games_avg = pd.DataFrame(round(lamb_2023[6:].agg({"Targets": "mean", "Rec_Yds": "mean"})))
display(first_6_games_avg)
last_6_games_avg = pd.DataFrame(round(lamb_2023[0:6].agg({"Targets": "mean", "Rec_Yds": "mean"})))
display(last_6_games_avg)
```

<h2>Now, I want to create regression plot to confirm the relationship between targets and receiving yards, which shows a significant positive correlation.<h2>

```python
sns.regplot(data=lamb_2023, 
            x="Targets", 
            y ="Rec_Yds", 
            ci=None,
            #scatter_kws={"alpha": 0.25}, 
            line_kws={"color": "navy"})
```

<h2>From the regression plot above, I can clearly see a difference after 9 targets, so I'm curious as to what the median is.  

```python
median_targets = np.median(lamb_2023["Targets"])
display(median_targets)
```

<h2>Now that I have confirmed the median number of targets is 9, I want to take quick look at the records that are equal to or less than the median.

```python
avg_yds_at_below_median = lamb_2023.query('Targets <= 9')
display(avg_yds_at_below_median)
```

<h2>I then want to take a look at the records that are above the median.

```python
avg_yds_at_below_median = lamb_2023.query('Targets > 9')
display(avg_yds_at_below_median)
```

<h2>I see them same impact when analyzing stats above median vs. at or below median as I did when splitting the stats into first 6 games vs. last 6 games.  Now, I want to create a new field, "Targets Above Median", to create categories that will be used for further analysis based on median targets. 

```python
lamb_2023["Targets_Above_Median"] = np.where(lamb_2023["Targets"] > median_targets, "yes", "no")
display(lamb_2023)
```

<h2>Utilizing this new field, I want to get a quick look at the average yards based on when targets were at or below median vs. above median.

```python
#avg_yds_by_median_targets = round(lamb_2023.groupby(["Targets_Above_Median"]).agg({"Rec_Yds": "mean"}),0)
avg_yds_by_median_targets = round(lamb_2023.groupby(["Targets_Above_Median"]).agg(Avg_Rec_Yds=("Rec_Yds", 'mean')),0)
avg_yds_by_median_targets = avg_yds_by_median_targets.reset_index()
display(avg_yds_by_median_targets)
```

<h2>Now, I want create bar chart to visualize the difference in performance between the two categories.

```python
color = ["Grey","Navy"]
sns.set_palette(color)

ax = sns.barplot(x="Targets_Above_Median", y="Avg_Rec_Yds", data=avg_yds_by_median_targets,           
                 width=.5, linewidth=2.5, dodge=False)

for p in ax.patches:    
    ax.annotate("%.2f" % p.get_height(), 
    (p.get_x() + p.get_width() / 2., p.get_height()),                 
    ha='center', 
    va='center', 
    fontsize=10, 
    color='gray', 
    xytext=(0, 5), 
    textcoords='offset points')
    
ax.set_title("Avg Yds by Median Targets")
```
